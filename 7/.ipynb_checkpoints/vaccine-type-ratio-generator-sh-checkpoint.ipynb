```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('cowin_vaccine_data_districtwise.csv',low_memory=False)
```

```python
data = data.rename(columns = data.iloc[0])
```

```python
data = data.drop(['S No','District_Key','Cowin Key'],axis=1)
```

```python
data1 = data[['State_Code','State','Covaxin (Doses Administered)','CoviShield (Doses Administered)']].copy()
```

```python
data1 = data1.iloc[1:,:]
```

```python
cols = []
count = 1
for column in data1.columns:
    if column == 'Covaxin (Doses Administered)':
        cols.append(f'Covaxin (Doses Administered){count}')
        count+=1
        continue
    cols.append(column)
data1.columns = cols
```

```python
colsa = []
count = 1
for column in data1.columns:
    if column == 'CoviShield (Doses Administered)':
        colsa.append(f'CoviShield (Doses Administered){count}')
        count+=1
        continue
    colsa.append(column)
data1.columns = colsa
```

```python
df2 = data1.filter(regex='Covaxin ')
df3 = data1.filter(regex='CoviShield ')
```

```python
newstuff = data1[['State_Code','State']].copy()
```

```python
val1 = pd.concat([newstuff, df2], axis=1)
val2 = pd.concat([newstuff, df3], axis=1)
```

```python
col_list= list(val1)
col_list.remove('State_Code')
col_list.remove('State')
# col_list
```

```python
def sum_frame_by_column(frame, new_col_name, list_of_cols_to_sum):
    frame[new_col_name] = frame[list_of_cols_to_sum].astype(float).sum(axis=1)
    return(frame)
```

```python
sum_frame_by_column(val1, 'covaxindoses', col_list)
```

```python
col_lista= list(val2)
col_lista.remove('State_Code')
col_lista.remove('State')
# col_lista
```

```python
sum_frame_by_column(val2, 'covishielddoses', col_lista)
```

```python
finalstuff1 = val1[['State_Code','State','covaxindoses']].copy()
finalstuff2 = val2[['State_Code','State','covishielddoses']].copy()
```

```python
result = pd.concat([finalstuff1, finalstuff2], axis=1, join='inner')
```

```python
newresult = result[['State_Code','State','covaxindoses','covishielddoses']].copy()
```

```python
newresult = newresult.loc[:,~newresult.columns.duplicated()]
```

```python
findf = newresult.groupby(['State_Code','State']).sum().reset_index()
res = findf.copy(deep=True)
```

```python
findf['vaccinationratio'] = findf['covishielddoses']/findf['covaxindoses']
```

```python
findf = findf.drop(['covaxindoses','covishielddoses'],axis=1)
```

```python
findf['State']=findf['State'].str.lower()
```

```python
stufffindf = findf[['State_Code','vaccinationratio']].copy()
```

```python
stufffindf.rename(columns = {'State_Code':'stateid'}, inplace = True)
```

```python
# stufffindf['vaccinationratio'] = stufffindf['vaccinationratio'].apply(np.floor)
```

```python
stufffindf.to_csv('vaccine-type-ratio-state.csv')
```
