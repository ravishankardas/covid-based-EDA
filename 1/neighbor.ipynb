```python
import pandas as pd
import re
```

```python
import json
with open("neighbor-districts.json",'r') as datafile:
    content = datafile.read()
    clean = content.replace('_', ' ').replace('district','').replace('/',' ').replace('Q','')    
    data = json.loads(clean)
```

```python
res = {key: [re.sub('\d', '', ele) for ele in val]
       for key, val in data.items()}
```
