```python
import pandas as pd
```

```python
df = pd.read_csv("dist_keys.csv")
```

```python
df.sort_values('District_Key')
```

```python
the_a = pd.read_csv("districts.csv")
# the_a.head()
```

```python
the_b = the_a[["Date","District","Confirmed"]].copy()
```

```python
date_df=the_b.loc[the_b['Date'] == "14-08-2021"]
date_df
```

```python
the_c = pd.read_csv("cowin_vaccine_data_districtwise.csv")
# the_c.head()
    
```

```python
the_d= the_c[["District_Key","District"]].copy()
the_d.sort_values('District_Key')
```

```python
int_df = pd.merge(df, the_d, how ='inner', on =['District_Key'])
tot_cases = pd.merge(int_df, the_b, how ='inner', on =['District'])
# int_df
```

```python

dict = {'District_Key': 'districtid',
        'District': 'timeid',
        'Confirmed': 'cases'}
tot_cases.rename(columns=dict,inplace=True)
tot_cases['timeid']='overall'
# tot_cases.to_csv("cases-overall.csv",index=None,)
```

```python
unique_districts=tot_cases['districtid'].unique()
cases_sum=[]
for i in unique_districts:
    cases_sum.append(tot_cases.loc[tot_cases['districtid'] == i, 'cases'].sum())
```

```python
df2=pd.DataFrame(unique_districts)
dict1 = {0: 'districtid'}
df2.rename(columns=dict1,inplace=True,)
# df2
```

```python
df1 = pd.DataFrame(cases_sum)
dict = {0: 'cases'}
df1.rename(columns=dict,inplace=True,)
# df1
```

```python
new_d=[]
txt='overall'
for i in range(273):
    new_d.append(txt)
df3 = pd.DataFrame(new_d)
dicta = {0: 'timeid'}
df3.rename(columns=dicta,inplace=True,)
# df3
```

```python
result = pd.concat([df2, df3], axis=1, join='inner')
final_result = pd.concat([result, df1], axis=1, join='inner')
```

```python
update_df = final_result.drop(38)
update_df.to_csv('cases-overall.csv')
update_df
```
