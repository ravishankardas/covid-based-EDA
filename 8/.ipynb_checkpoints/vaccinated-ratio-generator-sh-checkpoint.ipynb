```python
import pandas as pd
```

```python
data = pd.read_csv('cowin_vaccine_data_districtwise.csv',low_memory=False)
```

```python
data = data.rename(columns = data.iloc[0])
```

```python
data = data.drop(['S No','District_Key','Cowin Key'],axis=1)
```

```python
data1 = data[['State_Code','State','First Dose Administered','Second Dose Administered']].copy()
```

```python
data1 = data1.iloc[1:,:]
```

```python
cols = []
count = 1
for column in data1.columns:
    if column == 'First Dose Administered':
        cols.append(f'First Dose Administered{count}')
        count+=1
        continue
    cols.append(column)
data1.columns = cols
```

```python
colsa = []
count = 1
for column in data1.columns:
    if column == 'Second Dose Administered':
        colsa.append(f'Second Dose Administered{count}')
        count+=1
        continue
    colsa.append(column)
data1.columns = colsa
```

```python
df2 = data1.filter(regex='First ')
df3 = data1.filter(regex='Second ')
```

```python
newstuff = data1[['State_Code','State']].copy()
```

```python
val1 = pd.concat([newstuff, df2], axis=1)
val2 = pd.concat([newstuff, df3], axis=1)
```

```python
col_list= list(val1)
col_list.remove('State_Code')
col_list.remove('State')
```

```python
def sum_frame_by_column(frame, new_col_name, list_of_cols_to_sum):
    frame[new_col_name] = frame[list_of_cols_to_sum].astype(float).sum(axis=1)
    return(frame)
```

```python
sum_frame_by_column(val1, 'firstdose', col_list)
```

```python
col_lista= list(val2)
col_lista.remove('State_Code')
col_lista.remove('State')
```

```python
sum_frame_by_column(val2, 'seconddose', col_lista)
```

```python
finalstuff1 = val1[['State_Code','State','firstdose']].copy()
finalstuff2 = val2[['State_Code','State','seconddose']].copy()
```

```python
result = pd.concat([finalstuff1, finalstuff2], axis=1, join='inner')
```

```python
newresult = result[['State_Code','State','firstdose','seconddose']].copy()
```

```python
newresult = newresult.loc[:,~newresult.columns.duplicated()]
```

```python
findf = newresult.groupby(['State_Code','State']).sum().reset_index()
res = findf.copy(deep=True)
```

```python
findf['State']=findf['State'].str.lower()
```

```python
newData = pd.read_excel('DDW_PCA0000_2011_Indiastatedist.xlsx')
```

```python
popdata = newData[['Level','Name','TOT_P']].copy()
```

```python
rslt_df = popdata[popdata['Level'] == 'STATE'].copy()
```

```python
rslt_df.drop_duplicates(subset ="Name",keep = 'first', inplace = True)
```

```python
rslt_df['Name']=rslt_df['Name'].str.lower()
```

```python
rslt_df.sort_values("Name", axis = 0, ascending = True,inplace = True)
```

```python
rslt_df.reset_index(inplace = True)
```

```python
newrslt_df = rslt_df[['Name','TOT_P']].copy()
```

```python
newrslt_df.rename(columns = {'Name':'State'}, inplace = True)
```

```python
int_df = pd.merge(findf, newrslt_df, how='inner', on=['State'])
```

```python
int_df['vaccinateddose1ratio'] = int_df['TOT_P']/int_df['firstdose']
int_df['vaccinateddose2ratio'] = int_df['TOT_P']/int_df['seconddose']
```

```python
int_df.rename(columns = {'State_Code':'stateid'}, inplace = True)
```

```python
int_df.sort_values("vaccinateddose1ratio", axis = 0, ascending = True,inplace = True)
```

```python
int_df.reset_index(inplace = True)
```

```python
sachinint_df = int_df[['stateid','vaccinateddose1ratio','vaccinateddose2ratio']].copy()
sachinint_df.to_csv('vaccinated-dose-ratio-state.csv')
```
